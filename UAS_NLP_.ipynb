```python
# prompt: lakukan cleaning data pada column full_text

import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove mentions (@username)
    text = re.sub(r'@\S+', '', text)
    # Remove hashtags (#hashtag)
    text = re.sub(r'#\S+', '', text)
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_full_text'] = df['full_text'].apply(clean_text)
df[['full_text', 'cleaned_full_text']]
```

# **Membuat dan Melatih Model**
model ini akan difine-tune untuk klasifikasi umum (klasifikasi sentimen) mengelompokkan tweet ini berdasarkan kategori lain
melakukan inferensi menggunakan model pretrained dari Hugging Face, tanpa memerlukan pelabelan tambahan.

*melatih model DistilBERT*

```python
import pandas as pd
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
import torch

# Memuat tokenizer dan model DistilBERT
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)

# Ensure all values in 'cleaned_full_text' are strings
df['cleaned_full_text'] = df['cleaned_full_text'].astype(str)

# Melakukan tokenisasi data untuk inference
inputs = tokenizer(df['cleaned_full_text'].tolist(), truncation=True, padding=True, return_tensors="pt")

# Cek bentuk tensor untuk memastikan data siap
inputs['input_ids'].shape, inputs['attention_mask'].shape
```

```python
# prompt: tampilkan hasil modeling tersebut dan jelaskan dengan bahasa indonesia

# Melakukan inferensi dengan model
with torch.no_grad():
    outputs = model(**inputs)

# Mendapatkan probabilitas prediksi
probabilities = torch.softmax(outputs.logits, dim=1)

# Mendapatkan label prediksi (0 atau 1)
predicted_labels = torch.argmax(probabilities, dim=1)

# Menambahkan kolom 'predicted_label' ke DataFrame
df['predicted_label'] = predicted_labels.tolist()

# Menampilkan DataFrame dengan hasil prediksi
print(df[['cleaned_full_text', 'predicted_label']])

# Analisis hasil
positive_count = df[df['predicted_label'] == 1].shape[0]  # Contoh: label 1 untuk sentimen positif
negative_count = df[df['predicted_label'] == 0].shape[0] # Contoh: label 0 untuk sentimen negatif

print(f"Jumlah Sentimen Positif: {positive_count}")
print(f"Jumlah Sentimen Negatif: {negative_count}")

# (Opsional) Hitung persentase
total_tweets = len(df)
positive_percentage = (positive_count / total_tweets) * 100
negative_percentage = (negative_count / total_tweets) * 100

print(f"Persentase Sentimen Positif: {positive_percentage:.2f}%")
print(f"Persentase Sentimen Negatif: {negative_percentage:.2f}%")
```

Melatih Model dan Menyimpan Model Terlatih

```python
import pandas as pd
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
import torch

# Memuat dataset dan memeriksa label
data = pd.read_csv("/content/label.csv")

# Menghapus nilai kosong dan memastikan semua teks adalah string
data = data.dropna(subset=['cleaned_full_text', 'predicted_label'])  # Pastikan label ada
data['cleaned_full_text'] = data['cleaned_full_text'].astype(str)
data['predicted_label'] = data['predicted_label'].astype(int)  # Pastikan label berupa integer

# Inisialisasi tokenizer dan model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)

# Tokenisasi data
encodings = tokenizer(data['cleaned_full_text'].tolist(), truncation=True, padding=True, return_tensors="pt")
labels = torch.tensor(data['predicted_label'].tolist())  # Mengambil label

# Membuat DataLoader
class TweetDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __getitem__(self, idx):
        item = {key: val[idx].clone().detach() for key, val in self.encodings.items()}  # Gunakan clone().detach()
        item['labels'] = self.labels[idx]  # Menggunakan 'labels' sebagai kunci
        return item
    def __len__(self):
        return len(self.labels)

dataset = TweetDataset(encodings, labels)

# Tentukan argumen untuk training
training_args = TrainingArguments(
    output_dir='./results',
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

# Melatih model
trainer.train()

# Simpan model setelah pelatihan
model.save_pretrained("./saved_model")
tokenizer.save_pretrained("./saved_model")
```

```python
from fastapi import FastAPI
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
import torch

app = FastAPI()

# Memuat model dan tokenizer yang telah dilatih
model_path = "./saved_model"
tokenizer = DistilBertTokenizer.from_pretrained(model_path)
model = DistilBertForSequenceClassification.from_pretrained(model_path)

@app.post("/predict/")
async def predict(text: str):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs)
    logits = outputs.logits
    prediction = torch.argmax(logits, dim=1).item()
    return {"prediction": prediction}
```
